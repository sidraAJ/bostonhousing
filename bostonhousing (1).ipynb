```python
x = df.data
y = df.target
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size =0.2)
```

```python
ytrain.shape
```

```python
xtrain.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression, Lasso
regressor = LinearRegression()
regressor.fit(xtrain, ytrain)
  
# predicting the test set results
y_pred = regressor.predict(xtest)
```

```python
y_pred
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
mae = mean_absolute_error(ytest, y_pred)
mse = mean_squared_error(ytest, y_pred)
rmse = np.sqrt(mean_squared_error(ytest, y_pred))
r2_square = r2_score(ytest,y_pred)
```

```python
metrics = {"MAE":mae, "MSE":mse, "RMSE":rmse, "r2_score":r2_square}
print(metrics)
```

```python
plt.scatter(ytest, y_pred, c = 'green')
plt.xlabel("Price: in $1000's")
plt.ylabel("Predicted value")
plt.title("True value vs predicted value : Linear Regression")
plt.show()
```

# Another Models

```python
lasso = Lasso(alpha=0.1)
lasso.fit(xtrain, ytrain)
y_pred_lasso = lasso.predict(xtest)
lasso_mse = mean_squared_error(ytest, y_pred_lasso)
print('LASSO Test MSE:', lasso_mse)
```

```python
r2_square = r2_score(ytest,y_pred_lasso)
print('r2_square:', r2_square)
```
